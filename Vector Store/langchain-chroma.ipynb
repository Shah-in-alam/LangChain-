```python
from langchain_openai import  OpenAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
from langchain_core.documents import Document

# Create langchain documents for IPL PLayers
doc1=Document(page_content="Virat Kohli is a one of the most successful cricket players in the ipl history. Known for his aggressive batting style and consistency, he has been a key player for Royal Challengers Bangalore (RCB) since the inception of the IPL in 2008.",
                 metadata={"player": "Virat Kohli", "team": "RCB"})

doc2=Document(page_content="MS Dhoni, the former captain of the Indian cricket team, is renowned for his calm demeanor and exceptional finishing skills. He has led Chennai Super Kings (CSK) to multiple IPL titles and is considered one of the best wicket-keeper batsmen in the world.",
                 metadata={"player": "MS Dhoni", "team": "CSK"})

doc3=Document(page_content="AB de Villiers, often referred to as 'Mr. 360' for his ability to play shots all around the ground, is a South African cricketer who has been a vital part of RCB's lineup. His innovative batting and versatility make him a fan favorite.",
                 metadata={"player": "AB de Villiers", "team": "RCB"})  

doc4=Document(page_content="Rohit Sharma, the captain of Mumbai Indians (MI), is known for his elegant batting style and ability to score big hundreds. Under his leadership, MI has won several IPL titles, making him one of the most successful captains in the league's history.",
                 metadata={"player": "Rohit Sharma", "team": "MI"})

doc5=Document(page_content="Jasprit Bumrah is one of the premier fast bowlers in international cricket and a key player for Mumbai Indians (MI) in the IPL. Known for his unique bowling action and ability to deliver yorkers at will, Bumrah has been instrumental in MI's success over the years.",
                 metadata={"player": "Jasprit Bumrah", "team": "MI"})

```

```python
docs=[doc1, doc2, doc3, doc4, doc5]
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

vector_store = FAISS.from_documents(
    documents=docs,
    embedding=embeddings,
    perisist_directory="./faiss_ipl_vectorstore",
    collection_name="ipl_players"

)

print("FAISS vectorstore created ✅")

```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=["metadatas", "documents", "metadatas"])
```

```python
# search documnets

vector_store.similarity_search(
    query ="Who amnong these are a bowler?",
    k=2
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query ="Who among these are a bowler?",
    k=2
)
```

```python
# metadata filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Chennai Super Kings"},
)
```

```python
# update documnets 
updated_doc1=Document(page_content="Jasprit Bumrah is a fast bowler from India who plays for the Mumbai Indians (MI) in the IPL. He is known for his unique bowling action and ability to deliver yorkers at will. Bumrah has been instrumental in MI's success over the years and is considered one of the premier fast bowlers in international cricket")

vector_store.update_documents(documents=updated_doc1,documnet_id="1")
```
