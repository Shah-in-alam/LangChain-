# Retriever

```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# initialize the Wikipedia retriever( set the language top_k)
retriever = WikipediaRetriever(
    language="en",
    top_k_results=5
)
```

```python
query = "China–Pakistan relations"
docs = retriever.invoke(query)
len(docs)

```

```python
print(type(docs))
print(len(docs))



```

```python
if not docs:
    print("No documents found.")
else:
    for i, doc in enumerate(docs, start=1):
        print("=" * 60)
        print(f"RESULT {i}")
        print("=" * 60)

        print("\nCONTENT:\n")
        print(doc.page_content[:2000])  # first 2000 chars

        print("\nMETADATA:\n")
        print(doc.metadata)
```

# Vector store retriever

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document
```

```python
# step 1 : my documnets source

documents =[
    Document(page_content="Langchain helps developers build LLM apllication easily."),
    Document(page_content="FAISS is a library for efficient similarity search and clustering of dense vectors."),
    Document(page_content="Wikipedia is a free online encyclopedia that anyone can edit."),
    Document(page_content="Python is a popular programming language for data science and machine learning.")
]
```

```python
#step 2: Initialize the embeddings model
embeddings = OpenAIEmbeddings()
# step 3: create the vector store
vector_store=FIASS.from_documents(
    documents =documents,
    embedding=embeddings, 
    collection_name="my_documents"
)
```

```python
# Step 4: Convert vectorstore into a retriever
retriever =vectorstore.as_retriever(
    serarch_kwargs={"k":2}
)
```

```python
query="What is FISSA used for?"
results = retriever.invoke(query)
```

```python
for i, doc in enumerate(results):
    print(f"\n------- Results {i+1}---------")
    print(doc.page_content)
    
```

```python
results=vectorstore.similarity_search(query, k=2)
```

```python
for i, doc in enumerate(results):
    print(f"\n------- Results {i+1}---------")
    print(doc.page_content)
```
